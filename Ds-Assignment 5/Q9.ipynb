```python
import matplotlib.pyplot as plt

# Data to plot
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["red", "green", "blue", "pink", "gray", "brown"]



# Plot
plt.pie(popuratity,  labels=languages, colors=colors, autopct='%1.1f%%', shadow=True, startangle=0)

plt.axis('equal')
plt.show()
```
