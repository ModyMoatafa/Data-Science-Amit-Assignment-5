```python
import matplotlib.pyplot as plt
# x axis values
x = [1,4,5,6,7]
# y axis values
y = [2,6,3,6,3]


plt.plot(x, y, color='red', linestyle='dotted', linewidth = 2, marker='o', markerfacecolor='green', markersize=10)
#Set the y-limits of the current axes.
plt.ylim(1,10)

plt.xlim(1,10)


plt.xlabel('x - axis')

plt.ylabel('y - axis')


plt.show()
```
